```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = { 'Ay': ['Ocak', 'Ocak', 'Ocak', 'Ocak',
           'Şubat','Şubat', 'Şubat', 'Şubat',
           'Mart', 'Mart', 'Mart', 'Mart',
           'Nisan','Nisan','Nisan','Nisan'],
    'F-İnşaat-X-E-CİRO VERİLERİ': [17, 20, 26, 29,
              23,25,33,35,
              29,28,37,41,
              33,44,49,53]}

df = pd.DataFrame(data)
def ridgeline_plot(data, category, value, palette='viridis'):
    categories = data[category].unique()
    pal = sns.color_palette(palette, len(categories))
    fig, axes = plt.subplots(len(categories), 1, figsize=(8, len(categories) * 2), sharex=True)

    for i, cat in enumerate(categories):
        sns.kdeplot(data[data[category] == cat][value], ax=axes[i], fill=True, color=pal[i])
        axes[i].set_ylabel(cat, rotation=0, labelpad=30, size='large')
        axes[i].grid(True)

    plt.tight_layout()
    plt.show()

ridgeline_plot(df, 'Ay', 'F-İnşaat-X-E-CİRO VERİLERİ')
```
